```python
alpha = 0.2
```

```python
df['EWMA'] = df['Passengers'].ewm(alpha=alpha, adjust = False).mean()
```

```python
df.plot();
```

```python
manual_ewma = []

for x in df['Passengers'].to_numpy():
  if len(manual_ewma)>0:
    xhat =  alpha * x + (1-alpha)*manual_ewma[-1]
  else:
    xhat = x

  manual_ewma.append(xhat)

df['Manual_EWMA'] = manual_ewma
```

```python
df.plot();
```

```python
df.drop('Manual_EWMA', axis = 1, inplace = True)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
ses = SimpleExpSmoothing(df['Passengers'])
```

```python
df.index
```

```python
df.index_freq = 'MS'
```

```python
ses = SimpleExpSmoothing(df['Passengers'],
                         initialization_method='legacy-heuristic')
```

```python
res = ses.fit(smoothing_level = alpha, optimized = False)
```

```python
res
```

```python
df['SES'] = res.predict(start = df.index[0], end = df.index[-1])
```

```python
np.allclose(df['SES'], res.fittedvalues)
```

```python
df.plot()
```

```python
Ntest = 12
train = df.iloc[:-Ntest]
test = df.iloc[-Ntest:]
```

```python
ses = SimpleExpSmoothing(df['Passengers'],
                         initialization_method= 'legacy-heuristic')

res = ses.fit()
```

```python
train_idx = df.index <= train.index[-1]
test_idx = df.index > train.index[-1]
```

```python
df.loc[train_idx, 'SESfitted'] = res.fittedvalues
df.loc[test_idx, 'SESfitted'] = res.forecast(Ntest)
```

```python
df[['Passengers', 'SESfitted']].plot();
```

```python
from statsmodels.tsa.holtwinters import Holt
```

```python
holt = Holt(df['Passengers'],
            initialization_method = 'legacy-heuristic')

res_h = holt.fit()
```

```python
df['Holt'] = res_h.fittedvalues
```

```python
df[['Passengers', 'Holt']].plot();
```

```python
holt = Holt(train['Passengers'],
            initialization_method = 'legacy-heuristic')

res_h = holt.fit()

df.loc[train_idx, 'Holtfitted'] = res_h.fittedvalues
```

```python
df.loc[test_idx, 'Holtfitted'] = res_h.forecast(Ntest)
```

```python
df[['Passengers', 'Holtfitted']].plot();
```

```python
from  statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
hw = ExponentialSmoothing(train['Passengers'],
                          initialization_method = 'legacy-heuristic',
                          trend = 'add', seasonal = 'add', seasonal_periods = 12)

res_hw = hw.fit()

df.loc[train_idx, 'Holtwinters'] = res_hw.fittedvalues
df.loc[test_idx, 'Holtwinters'] = res_hw.forecast(Ntest)
```

```python
df[['Passengers', 'Holtwinters']].plot();
```

```python
def rmse(t, y):
  return np.sqrt(np.mean(y-t)**2)
```

```python
print("Train RMSE : ", rmse(train['Passengers'], res_hw.fittedvalues))
print("Test RMSE : ", rmse(test['Passengers'], res_hw.forecast(Ntest)))
```

```python
hw = ExponentialSmoothing(train['Passengers'],
                          initialization_method = 'legacy-heuristic',
                          trend = 'add', seasonal = 'mul', seasonal_periods = 12)

res_hw = hw.fit()

df.loc[train_idx, 'Holtwinters'] = res_hw.fittedvalues
df.loc[test_idx, 'Holtwinters'] = res_hw.forecast(Ntest)
```
